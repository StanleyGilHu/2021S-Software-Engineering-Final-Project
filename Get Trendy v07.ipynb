```python
import os
import numpy as np
import pandas as pd
from os import listdir
import re #used for regex findall
import collections
from collections import Counter
from os import walk

onlyfiles = [f for f in range(2006,2021)]
file_Name = ""
path_Name ="" #leads to the path/directory the file is located in
df1 = pd.DataFrame()
df2 = pd.DataFrame()
total_rows = 0
songWords = 0
gene_Name = ""
aveUniWords = 0
aveTotWords = 0



def dfFromCSV(path_Name, file_Name):
    #filename should be , ""'.', 'txt')CountWords.csv", or CountWords.csv if you 
    #openfile = open((path_Name+"\\"+file_Name),  encoding = "utf8")
    df1 = pd.read_csv(path_Name+"\\"+file_Name)
    return df1

def uniqueWordCount(df1):
    index=df1.index
    total_rows = len(index)
    #total_rows = df1['Word'].count + 1
    return total_rows
    #returns an integer of the unique words that appear in the song.
    
def totalWordCount(df1):  
    #songWords = df1.sum(axis = 1, skipna = True)
    songWords = df1['Count'].sum()
    return songWords
    #returns an integer of the total number of words that appear in a song.
    
def file_Process (dir_Name):
    aveUniWords = 0
    aveTotWords = 0
    onlyfiles2 = [f for f in listdir(dir_Name)]
    onlyfiles2 = filter(lambda x: x[-14:] == 'CountWords.csv', onlyfiles2)
    for f in onlyfiles2:
        df2 = dfFromCSV(dir_Name, f)
        aveUniWords = aveUniWords + uniqueWordCount(df2)
        aveTotWords = aveTotWords + totalWordCount(df2)
        
    aveTotWords = aveTotWords/100
    print(aveTotWords)
    print(" is the number of average total words per song in the year "+dir_Name)
    aveUniWords = aveUniWords/100
    print(aveUniWords)
    print(" is the number of average unique words per song in the year "+dir_Name)
    
for f in onlyfiles:
    path_Name = str(os.path.abspath(os.getcwd())+"\\"+str(f))
    #Where str f is the folders named by year.
    #print( path_Name )
    print("")

    file_Process(path_Name)
    
    
    
```
